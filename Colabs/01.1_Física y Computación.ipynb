
#Física y computación




<p align = "justify">La física computacional puede ser considerada desde diferentes perspectivas. Una, que es interesante para reflexionar sobre qué tipos de problemas de física es posible resolver, es interpretar cualquier fenómeno físico como el resultado de un cómputo. Desde este enfoque, hay tres elementos principales en un problema de física: un estado inicial, la evolución del sistema y el estado final que se mide. En el contexto de la computación clásica, estos corresponden a la entrada (input), que se codifica como una cadena de bits (0 o 1), una serie de compuertas lógicas (el cálculo) y la salida, que también se codifica como una cadena de bits. <p>

![Física como un cómputo](https://drive.google.com/uc?export=view&id=1dy4BTSjViOVuVkhFkQ4bqifcxh-yBkaz)

<p align = "justify">Nos podemos preguntar si esta perspectiva va más allá de ser una simple analogía y si es posible resolver cualquier problema de física usando una computadora clásica digital, dado que describimos al mundo de manera analógica y cuántica. La respuesta es que sí es posible (al menos en teoría). Las computadoras que utilizamos son computadoras universales, lo que significa que, con suficiente tiempo y memoria, pueden resolver cualquier problema con la precisión que se desee (siempre y cuando este problema no sea intrínsecamente insoluble). Si un problema tiene solución, se puede resolver con una computadora clásica.
En la práctica, el uso de una computadora clásica puede resultar ineficiente para ciertos problemas, como los que involucran sistemas cuánticos complejos. Por eso se están desarrollando nuevas técnicas de computación cuántica que pueden ser más eficientes para resolver este tipo de problemas. Esto se basa en la idea original de Feynmann de realizar simulaciones cuánticas (usar un sistema cuántico que podemos controlar con precisión para simular otro). Esta es una forma de computación en la que las leyes de evolución son cuánticas. Además, al utilizar fenómenos como la interferencia cuántica, es posible resolver algunos problemas de una manera más eficiente que con cualquier algoritmo clásico conocido. Por ejemplo encontrar los factores primos de un número usando el <a href=https://es.wikipedia.org/wiki/Algoritmo_de_Shor>algorithmo de Shor</a>.<p>







#Computación y Física



Más allá de la analogía entre física y computación, la computación es una herramienta esencial en la física moderna.

Se la usa en:

- Simulaciones: se puede codificar un sistema físico, como partículas e interacciones, y aplicar la dinámica del sistema (ecuaciones de movimiento) para calcular propiedades del sistema, como la presión y la temperatura.
- Resolución de sistemas de ecuaciones: por ejemplo las ecuaciones de Maxwell con ciertas condiciones de contorno.
- Problemas de autovalores y autovectores: estos problemas son importantes en muchos aspectos de la física, incluyendo la mecánica cuántica.
- Problemas de optimización: minimización de la energía o la búsqueda de soluciones.
- Cálculos analíticos con librerías que permiten hacer cálculos simbólicos.
- Minería de datos: es el proceso de extraer información útil de grandes volúmenes de datos, utilizando métodos de inteligencia artificial o aprendizaje automático. Puede ayudar a descubrir patrones y relaciones en los datos experimentales, teóricos o simulados.
- Adquisición de datos: la computación también se utiliza para adquirir y procesar datos experimentales.
- Procesamiento de datos experimentales: los datos obtenidos a partir de experimentos pueden ser analizados utilizando herramientas computacionales para extraer información relevante y hacer predicciones.

<p align = "justify">Las computadoras permiten ampliar enormemente la variedad de sistemas que se pueden estudiar (las soluciones analíticas son muy poco frecuentes). La posibilidad de usar la computadora como un laboratorio numérico y de visualizar las soluciones permite desarrollar la intuición física. Hace posible acceder a soluciones de modelos que capturan la física interesante o universal de un problema. Esos modelos son muchas veces demasiado complicados para un cálculo analítico pero omiten detalles que no son esenciales para entender el problema. <p>

<p align = "justify">
La física computacional tiene un papel crucial en la retroalimentación entre la física experimental y la física teórica. La física teórica se basa muchas veces en supuestos y modelos que no siempre se corresponden con la realidad experimental. Por otro lado, la física experimental genera datos que pueden ser difíciles de interpretar sin la ayuda de herramientas teóricas.
<p>
<p align = "justify">
La física computacional puede llenar esta brecha. Los modelos teóricos pueden ser simulados y comparados con los datos experimentales para verificar su validez, analizar las aproximaciones realizadas y hacer predicciones para futuros experimentos. Por otro lado, los datos experimentales pueden ser analizados con la ayuda de herramientas computacionales para extraer información relevante y compararla con las predicciones teóricas. <p>

Cada vez es más frecuente la reutilización de códigos. Si un grupo de investigadores desarrolla un código numérico y lo documenta adecuadamente, muchas veces distribuye una versión del código (por ejemplo en GitHub) y publica un artículo explicándolo. Otros investigadores pueden usar, y en general modificar el código, siempre y cuando citen la fuente.

Los beneficios de la reutilización de códigos son claros:
- Ahorro de tiempo y recursos: No hay que "reinventar la rueda" cada vez que se necesita resolver un problema.

- Replicabilidad: otros pueden usarlo para replicar resultados y controlar que sean correctos.

- Transparencia: permite una mejor comprensión y transparencia de los métodos y técnicas utilizados.

- Colaboración: se incentiva la colaboración entre investigadores y el avance de la ciencia en general.

- Impacto: Los investigadores que comparten sus códigos muchas veces obtienen un mayor reconocimiento y citas.
