# Guía para la conexión y uso del Cluster de Computadoras

La Gerencia de Física del CAB cuenta con una guía bastante detallada del uso del cluster de cálculo, la cual se puede consultar en el siguiente link:

https://fisica.cab.cnea.gov.ar/utilizacion

## Acceso al Cluster

Para poder acceder al cluster debe estar en el CAB o tener acceso VPN al CAB y un cliente ssh. Casi la totalidad de las distribuciones LINUX ya lo tienen preinstalado.

Si éste es su caso pude ejecutar el siguiente comando desde una terminal:

     ssh USUARIO@10.73.25.223

En caso que quisiera iniciar lo conexión desde un WINDOWS tendrá que instalar el cliente ssh que más le guste (openssh incluido en optional features de windows, putty, moabxterm, wsl, etc)

El nombre de USUARIO ha sido comunicado en el mail de confirmación de creación de la cuenta y será su nombre de usuario dentro de los clusters.

## Interface Web

Además, el cluster cuenta con una interface web a través de JupyterLab que funciona de una forma muy similar a como funciona `Google Colaboratory`. Para lo cual debemos conectarnos desde el siguiente link:

 https://login.fisica.cabib

 Al cargar la página nos va a solicitar la clave de acceso para poder entrar al cluster. Después, debemos ir a -> "Interactive Apps" -> "Jupyter Lab". Esto nos lleva a un formulario que envía una corrida con `Jupyter` solicitando tiempo de cálculo en uno de los Nodos del cluster.

 Las colas actuales más usadas son `copahue` y `gpu`.
