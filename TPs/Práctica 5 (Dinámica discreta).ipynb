##Ejercicio 3

El autómata celular más extraordinario de todos, y el más estudiado, es el *Juego de la Vida*, inventado por el matemático inglés John Conway. Consiste en una red cuadrada bidimensional, con condiciones de contorno periódicas, en la que cada nodo puede estar en uno de dos estados: activo o inactivo. El estado de cada nodo se actualiza en cada iteración, de acuerdo al número de vecinos activos en su vecindario de Moore (que incluye 8 vecinos), siguiendo las siguientes reglas:

* Si un nodo activo tiene menos de 2 vecinos activos, se inactiva.

* Si un nodo activo tiene más de 3 vecinos activos, se inactiva.

* Si un nodo inactivo tiene 3 vecinos activos, se activa.

* Si un nodo tiene 2 vecinos activos, no cambia.

Este autómata puede producir patrones que se preservan en el tiempo, que se mueven, que interactúan de maneras interesantes, e incluso que pueden replicarse (puede ver ejemplos en [este artículo de Wikipedia](https://en.wikipedia.org/wiki/Conway's_Game_of_Life)). Además, es capaz de realizar computación universal en el sentido de Turing.

Programe un Juego de la Vida y córralo a partir de condiciones iniciales aleatorias, para observar los patrones que emergen de manera espontánea. El sitio https://www.conwaylife.com es un excelente recurso mantenido por una comunidad de usuarios con muchísima información sobre la dinámica del Juego de la Vida, y con implementaciones que puede usar para comparar con la suya, o para inspirarse. Hay muchísima más información en la web, incluso implementaciones interactivas para correr en el navegador. En un capítulo de [Numberphile](https://www.youtube.com/watch?v=R9Plq-D1gEk), John Conway cuenta la historia de cómo inventó el Juego de la Vida. El artículo resultante, publicado en Scientific American, se convirtió en el éxito más grande de los más de 100 años de la revista.
